Beta distribution: 

The beta distribution is a continuous distribution which can take values between 0 and 1. This distribution is parameterized by two shape parameters α and β.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
import seaborn as sns

from scipy import stats

a = 0.5
b = 0.5
x = np.arange(0.01,1,0.01)
y = stats.beta.pdf(x,a,b)
plt.plot(x,y)
plt.title('Beta: a=%.1f,b=%.1f' %(a,b))

plt.xlabel('x')
plt.ylabel('Probability density')
plt.show()
```

One reason the beta distribution is interesting is because it is bounded to the [0, 1] interval (and therefor could be scaled or shifted to any finite interval), unlike many probability distributions that are unbounded. Beta Distribution is parametrized by two positive shape parameters, denoted by α and β, that appear as exponents of the random variable and control the shape of the distribution

**Applications**

The Beta distribution is used in a range of disciplines including rule of succession, Bayesian statistics, and task duration modeling. Examples of events that may be modeled by Beta distribution include:
1. The time it takes to complete a task
2. The proportion of defective items in a shipment
